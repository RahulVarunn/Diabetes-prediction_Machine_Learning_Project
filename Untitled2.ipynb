```python
def checking_dataset(dataset):
    print("Shape Of DataSet Rows: {} and Column: {}".format(dataset.shape[0],dataset.shape[1]))
    print("\nMean Of Dataset:\n\n{} ".format(dataset.mean()))
    print("\nMedian Of Dataset:\n\n{} ".format(dataset.median()))
    print("\nDataType Of Dataset:\n\n{} ".format(dataset.dtypes))
    print("\nNull Values Of Dataset:\n\n{} ".format(dataset.isnull().sum()))
    print("\nColumns Of Dataset:\n\n{} ".format(dataset.columns.values))
checking_dataset(dataset)
    
```

```python
def descriptive_stats(df):
    cmap1 = "gist_gray"
    desc = df.describe().T
    desc_df = pd.DataFrame(index = df. columns,columns = desc.columns,data = desc)
    f, ax = plt.subplots(figsize = (18, 8))
    
    sns.heatmap(desc,annot = True,fmt = ".2f",ax = ax,linecolor = "black",linewidths = 1,cbar = False,annot_kws = {"size" : 15})
    plt.yticks(size = 15, rotation = 0)
    plt.title("Descriptive Statistics", size = 15)
    plt.show()
descriptive_stats(dataset)
```

```python
#train  test split

X=dataset.drop("Outcome",axis=1)#dependent
Y=dataset['Outcome']#independent

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#training model 
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
#preduction
preduct=model.predict(X_test)
preduct
```

```python
acu=accuracy_score(preduct,Y_test)
print(acu)
```

```python
def result(i1,i2,i3,i4,i5,i6,i7,i8):
    resul_pred=model.predict([[i1,i2,i3,i4,i5,i6,i7,i8]])
    # print(resul_pred)
    if resul_pred==[1]:
        print("The Patient Is Suffering From Diabetes")
    else:
        print("The Patient Is NOT Suffering From Diabetes")

result(6 ,102 ,72 ,35 ,0 ,33.6 ,0.627 , 50 )


    
```
